# Collect Metrics

```python
def collect_ids_and_metrics(data_dict, metric):
    
    """
    Collects metrics for each VariantID across dfs collected in the dictionary
    
 Args: 
    data_dict: A dictionary, which keys (sorted) are names of ClinVar vcfs 
    and items are pandas dfs with nessesary columns for the analysis 
    metric: A column (CLNSIG or CLNREVSTAT) which should be summaerized per each VariantID across all used vcfs
    
  Returns:
     A dictionary where keys are variant IDs. Values are all CLNSIG or CLNREVSTAT records
    
    """
    
    all_ids = set()
    metrics_dict = {}
    
    for df in data_dict.values():
         all_ids.update(set(df['ID'].unique()))

        
    for year, df in data_dict.items():
        print(year)
        for id in all_ids:
            if id in df['ID'].to_numpy():
                metric_value = df[df['ID'] == id][f'{metric}'].tolist()[0]
                metrics_dict[id] = metrics_dict.get(id, []) + [metric_value]
            else:
                metrics_dict[id] = metrics_dict.get(id, []) + ['NA']
    
    return metrics_dict
```

# Collect COI * variants

*genetic variants with conflicting interpretations of pathogenicity

```python
def filter_conflicting_records(clnsig_data, coi_list):
  
  """
  Filters a ClinVar data dictionary to include only records with conflicting interpretations.

  Args:
      clnsig_data: A dictionary where keys are variant IDs and values are sorted lists of clinical significance records.
      conflicts_clnsig_records: A list of strings representing conflicting interpretation records.

  Returns:
      A dictionary containing only variants with at least one record from conflicts_clnsig_records. 
      The value for each variant remains the original list of clinical significance records.
  """
  
  flt_dict = {}
  for variant_id, clnsig_records in clnsig_data.items():
     for record in clnsig_records:
        if record in coi_list:
            flt_dict[variant_id] = clnsig_records  # Include entire record list
            break  # Stop iterating after finding a conflicting record
  
  return flt_dict
```
